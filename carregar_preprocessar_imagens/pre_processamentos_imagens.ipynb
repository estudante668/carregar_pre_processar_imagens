```python
import numpy as np
import os
import PIL
import PIL.Image
import tensorflow as tf
import tensorflow_datasets as tfds
```

```python
import pathlib

dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file('flower_photos.tgz',origin=dataset_url,
                                  extract=True)

PATH = os.path.join( data_dir, 'flower_photos')

# Listar subpastas (classes)
classes = [item for item in os.listdir(PATH) if os.path.isdir(os.path.join(PATH, item))]
print("Classes encontradas:", classes)

dados_flores = pathlib.Path(PATH)

```

```python
totalImagens = len(list(dados_flores.glob('*/*.jpg')))
print(totalImagens)
```
